```python
votes_average = movies_metadata['vote_average'].astype(int)
votes_count = movies_metadata['vote_count'].astype(int)
C = votes_average.mean()
m = votes_count.quantile(0.75)

def weighted_rating(x):
    v = x['vote_count']
    R = x['vote_average']
    return (v/(v+m) * R) + (m/(m+v) * C)

movies_metadata['weighted_rating'] = movies_metadata.apply(weighted_rating, axis=1)
```

```python
#Collaborative filtering based on user ratings

from scipy import sparse
movie_mat = movie_mat.fillna(0)
```

```python
def standardize(row):
    new_row = (row - row.mean())/(row.max()-row.min())
    return new_row

movie_mat_std = movie_mat.apply(standardize)
```

```python
movie_mat_std.head()
```

```python
# transpose is taken since we want similarity between items which need to be in rows
from sklearn.metrics.pairwise import cosine_similarity
item_sim = cosine_similarity(movie_mat_std.T)
item_sim_df = pd.DataFrame(item_sim, index = movie_mat_std.columns, columns = movie_mat_std.columns)

def get_similar_movies(movie_name, user_ratings):
    sim_score = item_sim_df[movie_name]*user_ratings
    sim_score = sim_score.sort_values(ascending = False)
    return sim_score
```

```python
#Content based recommender system
import sklearn
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfid = TfidfVectorizer(analyzer = 'word', ngram_range = (1,3), min_df = 0, stop_words='english')
movies_metadata['overview'] = movies_metadata['overview'].fillna('')
```

```python
tfid_matrix = tfid.fit_transform(movies_metadata['overview'])
tfid_matrix.shape
```

```python
from sklearn.metrics.pairwise import linear_kernel
cosine_sim = linear_kernel(tfid_matrix[:30000], tfid_matrix[0:30000])
```

```python
movies_metadata_1 = movies_metadata[:30000]
```

```python
indices = pd.Series(movies_metadata_1.index, index=movies_metadata_1['title'])
```

```python
indices[:10]
```

```python
def recommender_sys(title, cosine_sim = cosine_sim):
    inx = indices[title]
    score = list(enumerate(cosine_sim[inx]))
    score = sorted(score, key = lambda x: x[1],reverse = True)
    score = score[:11]
    movie_indices = [i[0] for i in score]
    return movies_metadata_1['title'].iloc[movie_indices]
```

```python
recommender_sys('Ratatouille')
```

```python
recommender_sys('Taare Zameen Par')
```
